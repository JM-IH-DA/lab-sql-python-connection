```python
# 2.

# Write a Python function called rentals_month that retrieves rental data for a given month 
# and year (passed as parameters) from the Sakila database as a Pandas DataFrame. 
# The function should take in three parameters: engine, month and year:
```

```python
# Function (rentals_month):

def rentals_month (engine, month, year):
    query=f"SELECT * FROM rental WHERE MONTH(rental_date)={month} AND YEAR(rental_date) = {year};"
    result=engine.execute(query)
    rentals_df = pd.DataFrame(result, columns=result.keys())
    return rentals_df
```

```python
# 3. 

# Develop a Python function called rental_count_month that takes the DataFrame provided by rentals_month 
# as input along with the month and year and returns a new DataFrame containing the number of rentals made 
# by each customer_id during the selected month and year.
```

```python
# In the context of the exercise, the rental_count_month function should accept the following parameters:

# rentals_df: DataFrame from the rentals_month function, which already contains data for the month and year.

# month: but the month cannot be other than the one that has been selected in the first function. Already defined.

# year: but the year cannot be other than the one that has been selected in the first function. Already defined.

# What I understand from the exercise is to use those 3 parameters for the new function, 

# but for me it does not make sense to use month and year with the rentals_month DF since 

# they are already defined in the first function.

# Anyway, here the function:
```

```python
# Function(rental_count_month 1):

def rental_count_month_1(rentals_df, month, year):
    rentals_df = rentals_df[(rentals_df['rental_date'].dt.month == month) & (rentals_df['rental_date'].dt.year == year)]
    rental_counts = rentals_df.groupby('customer_id').size().reset_index(name=f'rentals_count_{month:02d}_{year}')
    return rental_counts
```

```python
rental_count_month_1(rentals_df_august,8, 2005)
```

```python
rental_count_month_1(rentals_df_june,6, 2005)
```

```python
# The function is extracting the count from the Dataframe created with the function rentals_month, but sine
# we already selected one month, we cannot use the function for the whole information from the table, only
# for the month selected on the DF from rentals_month.
# For the previous rental_count_month function would be the same as:
```

```python
# Function(rental_count_month 2):

def rental_count_month_2(rentals_df):
    month = rentals_df['rental_date'].dt.month.values[0]
    year = rentals_df['rental_date'].dt.year.values[0]
    rental_counts = rentals_df.groupby('customer_id').size().reset_index()
    rental_counts.rename(columns={0: f'rentals_count_{month:02d}_{year}'}, inplace=True)
    return rental_counts
```

```python
rental_count_month_2(rentals_df_august)
```

```python
rental_count_month_2(rentals_df_june)
```

```python
# For me would have more sense to make a function where we can obtain the total rental counts for the selected
# month and year from all the information available on the table.
# Here the general function:
```

```python
# Function (rental_count_month 3):

def rental_count_month_3(engine, month, year):
    query=f"SELECT * FROM rental WHERE MONTH(rental_date)={month} AND YEAR(rental_date) = {year};"
    result=engine.execute(query)
    rentals_df = pd.DataFrame(result, columns=result.keys())
    rental_df_counts = rentals_df.groupby('customer_id').size().reset_index(name=f'rentals_counts_{month:02d}_{year}')
    return rental_df_counts
```

```python
rental_count_month_3(engine, 6, 2005)
```

```python
# 4. Create a Python function called compare_rentals that takes two DataFrames as input 
# containing the number of rentals made by each customer in different months and years. 
# The function should return a combined DataFrame with a new 'difference' column, 
# which is the difference between the number of rentals in the two months.
```

```python
# Here we can interpret the exercise in two ways:

# 1. We take two DF from the function: 'rental_count_month' with the months already defined, so we only have limited results
# and every time we want to compare different dates, we have to define our two DF first.

# Here the first function:
```

```python
# Defining first the 2 DF:
```

```python
df_1=rental_count_month_1(rentals_df_august, 8, 2005)
df_2=rental_count_month_1(rentals_df_june, 6, 2005)
display(df_1.head())
display(df_2.head())
```

```python
def compare_rentals(df1, df2):
    merged_df = pd.merge(df1, df2, on='customer_id', how='outer')
    merged_df.fillna(0, inplace=True)
    merged_df['difference'] = merged_df.iloc[:, 1] - merged_df.iloc[:, 2]
    merged_df['difference']= merged_df['difference'].abs()
    return merged_df
```

```python
compare_rentals(df_1,df_2)
```

```python
# The previous DF only shows the results for the defined DF.
```

```python
# The second way, and the most general and logical, would be as follows:
```

```python
def compare_rentals(engine, month1, year1, month2, year2):
    
    query1 = f"SELECT customer_id, COUNT(rental_id) as rentals_{month1:02d}_{year1} FROM rental WHERE MONTH(rental_date) = {month1} AND YEAR(rental_date) = {year1} GROUP BY customer_id"
    result1 = engine.execute(query1)
    df1 = pd.DataFrame(result1, columns=result1.keys())

    query2 = f"SELECT customer_id, COUNT(rental_id) as rentals_{month2:02d}_{year2} FROM rental WHERE MONTH(rental_date) = {month2} AND YEAR(rental_date) = {year2} GROUP BY customer_id"
    result2 = engine.execute(query2)
    df2 = pd.DataFrame(result2, columns=result2.keys())

    
    merged_df = pd.merge(df1, df2, on='customer_id', how='outer')
    
    merged_df.fillna(0, inplace=True)
    
    merged_df['difference'] = merged_df[f'rentals_{month1:02d}_{year1}'] - merged_df[f'rentals_{month2:02d}_{year2}']
    
    merged_df['difference'] = merged_df['difference'].abs()
    
    return merged_df
```

```python
# With this function we can obtain directly the info from all the table (absolute values):
```

```python
compare_rentals(engine, 6, 2005, 8, 2005)
```

```python
compare_rentals(engine, 8, 2005, 6, 2005)
```
